```python
xx = pd.DataFrame(index = ['aa', 'bb', 'cc', 'dd', 'ee'], 
                  data={'Apples':[20,10,15,25,18],
                        'Oranges':[12,22,17,24,20]})
```

```python
xx
```

```python
ax = xx.plot.bar(figsize=(15,4), color =['0.8','0.2'])
```

```python
ax = xx.plot.kde(figsize=(15,4))
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,4))
fig.suptitle('Two Variable Plots')
xx.plot.line(ax=ax1)
xx.plot.scatter(ax=ax2, x='Apples', y='Oranges', title='Scatter plot')
xx.plot.bar(ax=ax3, title='Bar plot')
```
